#### Solution

````python
employee_data = {
    "E001": {"name": "Alice", "hours_worked": 40, "sales_made": 10},
    "E002": {"name": "Bob", "hours_worked": 35, "sales_made": 12},
    "E003": {"name": "Charlie", "hours_worked": 45, "sales_made": 15},
    "E004": {"name": "Diana", "hours_worked": 50, "sales_made": 25},
}

# Total hours worked and total sales made (2.5 points each)
total_hours = sum(employee["hours_worked"] for employee in employee_data.values())
total_sales = sum(employee["sales_made"] for employee in employee_data.values()) 

# Employee with highest sales/hours ratio (2 options - either option receives full credit) (5 points)
# 1
top_performer = max(employee_data.items(), key=lambda employee: employee[1]["sales_made"] / employee[1]["hours_worked"])
top_performer = top_performer[1]['name']

# 2
top_performer = ''
top_ratio = 0
for id, emp in employee_data.items():
    ratio = emp['sales_made'] / emp['hours_worked']
    if ratio > top_ratio:
        top_ratio = ratio
        top_performer = emp['name']

# Employees who have worked more than 40 hours (3 points)
overworked_employees = [data["name"] for data in employee_data.values() if data["hours_worked"] > 40]

print(f"Total Hours Worked: {total_hours}")
print(f"Total Sales Made: {total_sales}")
print(f"Top Performer: {top_performer}")
print("Employees with Over 40 Hours:")
for name in overworked_employees:
    print(f"-- {name}")

################## Rubric (Total: 15 points)
# Correctly determining total hours worked and total sales (5 points)
# Correctly identifying the employee with the highest sales to hours worked ratio (5 points)
# Correctly listing employees who have worked more than 40 hours (3 points)
# Code readability  (2 points)
# Additional Rubrics (PLEASE READ):
# No point deduction for missing COMMENTS
# a) Do not deduct points if variable names are not meaningful OR if 'for in range()' is used
# b) If there are minor issues, such as syntax errors in the entire program, deduct only 2 points total for the entire program
# c) Give at least total 6 points if the code is partially okay, but not correct or close to correct. This is only if you didn't apply other rubrics.
# d) If students miss one/two spaces to show the alignment, there is no point deduction.
# e) In general, code should be mostly in python format but pseudocode is fine as long as it is algorithmically correct
````

````python
inventory_data = {
    "Smartphone": {"category": "electronics", "price": 799.99, "quantity": 5},
    "Table": {"category": "furniture", "price": 349.99, "quantity": 2},
    "Laptop": {"category": "electronics", "price": 99.99, "quantity": 10},
    "Bed": {"category": "furniture", "price": 549.99, "quantity": 3},
    "Dress": {"category": "clothing", "price": 24.99, "quantity": 50},
}

# Calculate total value of items in stock for each category
total_values = {}
for item, details in inventory_data.items():
    category = details['category']
    total_value = details['price'] * details['quantity']
    total_values[category] = total_values.get(category, 0) + total_value

# Find item with the highest quantity in stock
#highest_quantity_item = max(, key=lambda key: inventory_data[key][????])
highest_qty = 0
highest_qty_item = ''
for item, details in inventory_data.items():
    if details['quantity'] > highest_qty:
        highest_qty = details['quantity']
        highest_qty_item = item


# Output names and prices of items that are priced above $250
# expensive_items = {item: details[????] for item, details in ???? if ????}
expensive_items = {}
for item, details in inventory_data.items():
    if details['price'] > 250:
        expensive_items[item] = details['price']

print("Total Value by Category:")
for category, total in total_values.items():
    print(f"- {category}: {total:.2f}")

print(f"Item with Highest Quantity: {highest_qty_item}")

print("Items Priced Above $250:")
for item, price in expensive_items.items():
    print(f"- {item}: {price:.2f}")
````
